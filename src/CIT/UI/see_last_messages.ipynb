```python
import os

import pandas as pd

from CIT.evaluation.utils import load_jsonl
```

```python
messages_folder="./src/CIT/UI/messages"
messages=[]
for filename in os.listdir(messages_folder):
    if filename.endswith(".jsonl"):
        if "Michel" in filename or "User" in filename:
            continue
        file_path = os.path.join(messages_folder, filename)
        user_messages = load_jsonl(file_path)
        for message in user_messages:
            user=filename.split('.')[0]
            message['user'] = user
        messages.extend(user_messages)
messages_df = pd.DataFrame(messages)
messages_df.timestamp = pd.to_datetime(messages_df.timestamp, format='mixed')
messages_df.sort_values(by="timestamp",ascending=False).head(10)
```
