```python

import matplotlib.pyplot as plt

from CIT.evaluation.utils import load_jsonl
```

```python
all_data = load_jsonl("../data/all_data.jsonl")
```

```python
retrieved_contexts = [len(d["retrieved_context"].split()) for d in all_data]
```

```python
plt.hist(retrieved_contexts, bins=100)
plt.xlabel("Number of words in retrieved contexts")
plt.ylabel("Frequency")
plt.title("Distribution of Retrieved Context Lengths")
plt.show()
```
