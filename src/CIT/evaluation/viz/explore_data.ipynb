# Notebook to view data length and train test split

```python
import json
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from ..utils import load_jsonl
```

```python
dir = "./src/CIT/documents/confluence_json_without_root_with_titles"
```

```python
files_content = []
for file in os.listdir(dir):
    with open(os.path.join(dir, file), "r") as f:
        file_dico = json.load(f)
    if "content" in file_dico:
        files_content.append(file_dico["content"])
    else:
        print(f"File {file} does not contain 'content' key.")
        continue
```

```python
print(f"Number of files: {len(files_content)}")
```

Length of documents

```python
dir_images = "./src/CIT/evaluation/viz/images"
```

```python
# plot length of documents in terms of nukmber of tokens and number of characters
lengths_char = []
lengths_tokens = []
for i in range(len(files_content)):
    lengths_char.append(len(files_content[i]))
    lengths_tokens.append(len(files_content[i].split()))
# Create a DataFrame
df = pd.DataFrame({"lengths_char": lengths_char, "lengths_tokens": lengths_tokens})
# create an hitsogram of the lengths on 2 plots
fig, axs = plt.subplots(2, 1, figsize=(6, 6))
# Plot histogram of lengths_char
axs[0].hist(df["lengths_char"], bins=50, color="blue", alpha=0.7)
# add median line
median_char = np.median(df["lengths_char"])
axs[0].axvline(median_char, color="red", linestyle="dashed", linewidth=1)
# add median text
axs[0].text(median_char + 400, 10, f"Median: {median_char}", color="red")
axs[0].set_title("Histogram of Document Lengths (Characters)")
axs[0].set_xlabel("Length (Characters)")
axs[0].set_ylabel("Frequency")
# axs[0].set_yscale('log')
# Plot histogram of lengths_tokens
axs[1].hist(df["lengths_tokens"], bins=50, color="green", alpha=0.7)
# add median line
median_tokens = np.median(df["lengths_tokens"])
axs[1].axvline(median_tokens, color="red", linestyle="dashed", linewidth=1)
# add median text
axs[1].text(median_tokens + 10, 10, f"Median: {median_tokens}", color="red")
axs[1].set_title("Histogram of Document Lengths (Tokens)")
axs[1].set_xlabel("Length (Tokens)")
axs[1].set_ylabel("Frequency")
# set ticks to integers
axs[0].set_xticks(np.arange(0, max(df["lengths_char"]), 1000))
axs[1].set_xticks(np.arange(0, max(df["lengths_tokens"]), 100))
axs[0].set_yticks(np.arange(0, 15, 2))
axs[1].set_yticks(np.arange(0, 17, 2))
# axs[1].set_xscale('log')
# Show the plots
plt.tight_layout()
plt.savefig(os.path.join(dir_images, "lengths_docs_CIT_histograms.png"))
plt.show()
```
