```python
import os

from CIT.evaluation.utils import load_jsonl, save_jsonl
```

```python
folder_all_base_answers= 'data/base_answers_citing'
answers= []
for file in os.listdir(folder_all_base_answers):
    if file.endswith('.jsonl'):
        file_path = os.path.join(folder_all_base_answers, file)
        answers.extend(load_jsonl(file_path))
```

```python
perfect_answers= []
for answer in answers:
    if answer['precision'] == 1 and answer['recall'] == 1:
        perfect_answers.append(answer)
    
bad_answers= []
for answer in answers:
    if answer['quality'] == 0:
        bad_answers.append(answer)
```

```python
print(f'Number of perfect answers: {len(perfect_answers)}')
print(f'Number of bad answers: {len(bad_answers)}')
```

```python
save_jsonl("data/perfect_answers_base.jsonl", perfect_answers)
```

```python
save_jsonl("data/bad_answers_base.jsonl", bad_answers)
```
