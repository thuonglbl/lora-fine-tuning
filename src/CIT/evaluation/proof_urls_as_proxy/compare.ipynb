# Check correlation

```python
#check correlation
from scipy.stats import pearsonr
```

```python
metric_res=[1 for i in range(406)]+[0 for i in range(59)]
human_scores=[1 for i in range(402)]+[0 for i in range(4)]+[1 for i in range(4)]\
    +[0 for i in range(53)]+[0.5 for i in range(2)]
```

```python
pearson_corr, p_value = pearsonr(metric_res, human_scores)
print(f"Pearson correlation: {pearson_corr}, p-value: {p_value}")
```
