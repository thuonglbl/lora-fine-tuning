```python
from IPython.display import Image, display
from langgraph.checkpoint.memory import MemorySaver
from langgraph.graph import END, MessagesState, StateGraph
```

```python
graph_builder = StateGraph(MessagesState)
graph_builder.add_node(chunkenization)
graph_builder.add_node(vectorization)
graph_builder.add_node(retrieval)
graph_builder.add_node(generation)
graph_builder.add_node(user_input)
graph_builder.add_node(optional_query_refinement)
# graph_builder.add_node(conversation_memory)

graph_builder.add_edge("chunkenization", "vectorization")
graph_builder.add_edge("vectorization", "retrieval")
graph_builder.add_edge("retrieval", "generation")
graph_builder.add_edge("user_input", "optional_query_refinement")
graph_builder.add_edge("optional_query_refinement", "retrieval")
# graph_builder.add_edge("conversation_memory","generation")


graph_builder.add_edge("generation", END)
graph_builder.set_entry_point("chunkenization")
graph = graph_builder.compile()
memory = MemorySaver()
graph = graph_builder.compile(checkpointer=memory)
display(Image(graph.get_graph().draw_mermaid_png()))
```
