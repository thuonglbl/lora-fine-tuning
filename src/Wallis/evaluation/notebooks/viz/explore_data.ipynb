# Check document lenghts

```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dir = "./transformed_documents"
```

```python
files_txt = []
for filename in os.listdir(dir):
    fullname = os.path.join(dir, filename)
    with open(fullname, "r") as f:
        content = f.read()
    files_txt.append(content)
```

```python
print(f"Number of files: {len(files_txt)}")
```

Length of documents

```python
dir_images = "./evaluation/notebooks/viz/images"
```

```python
# plot length of documents in terms of nukmber of tokens and number of characters
lengths_char = []
lengths_tokens = []
for i in range(len(files_txt)):
    lengths_char.append(len(files_txt[i]))
    lengths_tokens.append(len(files_txt[i].split()))
# Create a DataFrame
df = pd.DataFrame({"lengths_char": lengths_char, "lengths_tokens": lengths_tokens})
# create an hitsogram of the lengths on 2 plots
fig, axs = plt.subplots(2, 1, figsize=(6, 6))
# Plot histogram of lengths_char
axs[0].hist(df["lengths_char"], bins=50, color="blue", alpha=0.7)
# add median line
median_char = np.median(df["lengths_char"])
axs[0].axvline(median_char, color="red", linestyle="dashed", linewidth=1)
# add median text
axs[0].text(median_char + 10000, 100, f"Median: {median_char}", color="red")
axs[0].set_title("Histogram of Document Lengths (Characters)")
axs[0].set_xlabel("Length (Characters)")
axs[0].set_ylabel("Frequency")
axs[0].set_yscale("log")
# Plot histogram of lengths_tokens
axs[1].hist(df["lengths_tokens"], bins=50, color="green", alpha=0.7)
# add median line
median_tokens = np.median(df["lengths_tokens"])
axs[1].axvline(median_tokens, color="red", linestyle="dashed", linewidth=1)
# add median text
axs[1].text(median_tokens + 1000, 100, f"Median: {median_tokens}", color="red")
axs[1].set_title("Histogram of Document Lengths (Tokens)")
axs[1].set_xlabel("Length (Tokens)")
axs[1].set_ylabel("Frequency")
axs[1].set_yscale("log")
# Show the plots
plt.tight_layout()
plt.savefig(os.path.join(dir_images, "lengths_docs_Valais_histograms.png"))
plt.show()
```
